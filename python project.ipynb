```python
#Q4

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

revenue_table = soup.find("table", class_="historical_data_table")

revenue_data = []
table_rows = revenue_table.find_all("tr")
for row in table_rows[1:]:
    data_cells = row.find_all("td")
    year = data_cells[0].text
    revenue = data_cells[1].text.replace("$", "").replace(",", "")
    revenue_data.append({"Year": year, "Revenue": float(revenue)})


gme_revenue = pd.DataFrame(revenue_data)

gme_revenue.tail()
```

```python
#Q5

import matplotlib.pyplot as plt

import yfinance as yf


tesla_data = yf.download("TSLA")



def make_graph(data, title):
    plt.plot("tesla_data")
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    plt.title(title)
    plt.show()

    
    

make_graph(tesla_data, 'Tesla Stock Price')
```
